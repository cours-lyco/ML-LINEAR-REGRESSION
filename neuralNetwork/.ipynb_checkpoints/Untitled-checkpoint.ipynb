```python
import sys, os
from collections import OrderedDict
import numpy as np 

'''path of external device '''
rootDir = "/Volumes/Disk E/"
datasDir = rootDir + "/crohme2016_inkml_datasets/CROHME2016_data/"

imagesDir = rootDir + "/crohme2016_inkml_images/"
'''now get train valid test dir'''

trainDir = datasDir + 'Task-2-Symbols/task2-trainSymb2014/'
trainDirDatasets = trainDir + 'trainingSymbols/'
trainDir_ground_truth = trainDir + 'trainingSymbols/iso_GT.txt'
trainDatasImmagesDir = "/Volumes/Disk E/crohme2016_inkml_images/train_images"

testDir = datasDir + 'Task-2-Symbols/task2-testSymbols2014/'
testDir_ground_truth = testDir + "testSymbols_2016_iso_GT.txt"
testDirDatasets = testDir + "testSymbols/"
testDatasImmagesDir = "/Volumes/Disk E/crohme2016_inkml_images/test_images"

valDir = datasDir + 'Task-2-Symbols/task2-validation-isolatedTest2013b/'
valDirDatasets = datasDir + 'Task-2-Symbols/task2-validation-isolatedTest2013b/'
valDir_ground_truth = valDirDatasets + 'iso_GT.txt'
valDatasImmagesDir = "/Volumes/Disk E/crohme2016_inkml_images/val_images"
```

```python
class GT(object):
   
    def __init__(self, name):  # name == get_train_data OR name == get_test_data OR name == get_val_data
      
        if 'get_train_data' == name :
         self.gt_abs_path = trainDir_ground_truth
         self.imagesDir = trainDatasImmagesDir
         self.gt = 'train'

      if 'get_test_data' == name :
         self.gt_abs_path = testDir_ground_truth
         self.imagesDir = testDatasImmagesDir
         self.gt = 'test'

      if 'get_val_data' == name :
         self.gt_abs_path = valDir_ground_truth
         self.imagesDir =  valDatasImmagesDir
         self.gt = 'validation'

      self.sortedDict = self.read_gt(self.gt_abs_path)
      #arr = self.create_to_categorical_array(self.sortedDict)
      
      #a = self.get_label_value_in_categorical_array('X', self.sortedDict, arr)
   
   #------------------------------------------------------------------------------#
   # return a Dictionary 
   # { 
   # ')' : [abs_path_image1, abs_path_imag2, ...]
   #    .....................................
   #   '!'  : [abs_path_image14, abs_path_imag267, ...]
   # }
   #------------------------------------------------------------------------------#
   def read_gt(self, abs_gt_file_path):
      '''Loads the routes of the png files'''
      try : 
         print("reading ground truth {} ..................".format(self.gt))
         with open(abs_gt_file_path) as file:
            lines = file.readlines()
      
            aDict = {}
            aLabels = []

            for line in lines:
               line = line.strip()
            
               parts = line.split(',')
               label = parts[1].strip()

               data = parts[0].split('_')
               index = data[-1].strip()
               
               if label not in aLabels and label != 'junk':
                  aLabels.append(label)

               
               if label in aDict and label != 'junk' :
                  aDict[label].append(self.imagesDir + '/' + 'iso' + index + '.png')
               else:
                  aDict[label] = [self.imagesDir + '/' + 'iso' + index + '.png']
         
         #sort dictionary but others method doesn't work
         bDict = {}
         for label in aLabels:
            bDict[label] = aDict[label]

         #aDict = OrderedDict(sorted(aDict.items(), key=lambda t: t[0]))
         # print("dictionnary sorted keys : ", bDict.keys())
         print("Finish readin ground {} truth: {}".format(self.gt, len(bDict)))
         #print("bDict length:........", len(bDict))
         return bDict
      
      except FileNotFoundError as e:
            print(e)

   def  create_to_categorical_array(self, sortedDict):
      length = len(sortedDict)
     
      matrixCategorical = np.zeros((length, length), dtype=np.float32)
     
      np.fill_diagonal(matrixCategorical, 1)
      
      #print("matrix diagonal :.........................\n", matrixCategorical)
      return  matrixCategorical

   def  get_label_value_in_categorical_array(self, label, sortedDict, categoricalArray):
         aKeys = list(sortedDict.keys())
         index = aKeys.index(str(label))
         
         indexInCategorical = categoricalArray[index][index]
         print("category arr: ", indexInCategorical)
         return indexInCategorical
```
